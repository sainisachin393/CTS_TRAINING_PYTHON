```python
sql_query=pd.read_sql_query('''select * from problem''',mydb)
```

```python
df1=pd.DataFrame(sql_query,columns=['Location','Period','Vals'])
```

```python
display(df1)
```

```python
sql_query_2=pd.read_sql_query('''select * from new1''',mydb)
```

```python
df2=pd.DataFrame(sql_query_2,columns=['START_PERIOD','END_PERIOD','PERIOD_RANGE'])
```

```python
display(df2)
```

```python
df3=pd.merge(df1,df2,right_index=True,left_index=True)
```

```python
df3
```

```python
df3.dtypes
```

```python
df3[['END_PERIOD']]=df3[['END_PERIOD']].apply(pd.to_numeric)
```

```python
#defining the threshold of our year groups
year_groups=[1990,1996,2000,2005,2010,2015,2020]
year_group_names=['before 1996','1996-2000','2001-2005','2006-2010','2011-2015','2016-2020']
```

```python
df3['year group']=pd.cut(df3['END_PERIOD'],bins=year_groups,labels=year_group_names)
```

```python
df3.head()
```

```python
df4=pd.pivot_table(df3,values='Vals',index='Location',columns='year group',aggfunc='mean')
```

```python
df4
```

```python
df5=df4.fillna('0')
```
