# Plotting


```python
sns.pairplot(df)
```

```python
# count plot for gender
sns.countplot(df.Gender);
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(x='Customer type',y='Tax 5%',data=df,hue='Payment')
```

```python
sns.distplot(df['gross income'],kde=True)
```

```python
plt.figure(figsize=(8,3))
sns.boxplot(y=df['Total'],x=df['Payment'],hue=df.Gender)
```

```python
sns.countplot(y='Product line',data=df)
```

```python
sns.lineplot(y='Product line',x='Quantity',data=df)
```

```python
sns.scatterplot(x='Branch',y='gross income',data=df)
```

```python
sns.violinplot(x='Branch',y='Total',data=df)
```

```python
sns.relplot(y='Product line',x="gross income",data=df)
```

```python
g = sns.FacetGrid(df, col = "Payment", hue = "Gender")
g.map(plt.scatter, "Total", "Unit price", alpha =.7)

g.add_legend();
```
