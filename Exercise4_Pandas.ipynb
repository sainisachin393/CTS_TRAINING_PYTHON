# Merge and Joining

```python
contacts = pd.DataFrame(
    {
        'name' : ['Alice', 'Charlie'],
        'tel'  : ['+44 1234 5678', '+44 3245 5564']
    }
)
```

```python
exam_results.merge(contacts, left_on='names', right_on='name') 
```

```python
exam_results.merge(contacts, left_on='names', right_on='name', how='outer')
```
